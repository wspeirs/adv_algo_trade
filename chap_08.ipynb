# 8.1.1 Example: Sample Covariance in R

````python
import polars as pl
import numpy as np

pl.set_random_seed(0)
generator = np.random.default_rng(0x0)
````

````python
x = np.arange(1, 101) + 20.0*generator.normal(size=100)
y = np.arange(1, 101) + 20.0*generator.normal(size=100)

np.cov([x, y])  # can also be np.cov(x, y)
````

Note that the values on the diagonal are the variances, and the off-diagonals are the covariances.

```
[[variance of X, covariance of X and Y],
 [covariance of X and Y, variance of Y]]
```

````python
np.corrcoef([x, y])
````

# 8.5 The Correlogram


````python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

w = generator.normal(size=100)

plot_acf(w)

plt.show()
````

````python
w_seq = np.arange(1, 101)
w_rep = np.tile(np.arange(1, 11), 10)

fig, axes = plt.subplots(2, 1)

plot_acf(w_seq, ax=axes[0])
plot_acf(w_rep, ax=axes[1])

plt.show()
````
